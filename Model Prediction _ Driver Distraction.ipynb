```python
from keras.models import model_from_json

# Model reconstruction from JSON file
with open('../input/distracted-driver-detection/model.json', 'r') as f:
    model = model_from_json(f.read())

# Load weights into the new model
model.load_weights('../input/distracted-driver-detection/weights.h5')
```

```python
img_path = "../input/state-farm-distracted-driver-detection/imgs/test/img_100148.jpg"
```

```python
import cv2
img_array = cv2.imread(img_path)
new_img = cv2.resize(img_array,(224,224))
```

```python
import matplotlib.pyplot as plt
plt.imshow(new_img)
plt.show()
```

```python
import numpy as np
test = np.array(new_img.reshape(-1,224,224,3))
```

```python
preds = model.predict(test)
```

```python
preds=np.argmax(preds)
```

```python
classes = {0: "safe driving",
1: "texting - right",
2: "talking on the phone - right",
3: "texting - left",
4: "talking on the phone - left",
5: "operating the radio",
6: "drinking",
7: "reaching behind",
8: "hair and makeup",
9: "talking to passenger",
}


for key,value in classes.items():
    if preds==key:
        predicted = value

predicted 
```
